```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()
```

```python
cols = ['fare', 'age', 'survived']
```

```python
arr = np.load('../data/arr15.npy')

df = pd.DataFrame(arr, columns = cols)

df.head(3)
```

```python
sns.distplot(df.fare)
```

```python
sns.lmplot('fare', 'age', df)
```

## 1. Normalize the data
Rescale it to the range 0-1

```python
def normalizr(arr):
    mn = arr.min()
    mx = arr.max()
    denom = mx-mn
    
    return (arr-mn)/denom
```

```python
arr_norm = np.load('../data/arr15.npy')

arr_norm[:,0] = normalizr(arr_norm[:,0])
arr_norm[:,1] = normalizr(arr_norm[:,1])
```

```python
df_norm = pd.DataFrame(arr_norm, columns = cols)

df_norm.head()
```

## 2. Validate they move the same

```python
sns.distplot(df.fare)
```

```python
sns.distplot(df_norm.fare)
```

```python
sns.lmplot('fare', 'age', df)
```

```python
sns.lmplot('fare', 'age', df_norm)
```
